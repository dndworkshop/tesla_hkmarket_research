```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import cycle, islice
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/csv/hk first registration number of cars.csv')
df
```

```python
df.drop(df.index[0:9], inplace=True)
df = df.drop(df.iloc[:, 7:17],axis = 1)
df
```

```python
col_names = ['Year', 'Month', 'Petrol', 'Electric', 'Diesel', 'Other(4)', 'Sub-total']
df.columns = col_names
df.drop(df.index[0:4], inplace=True)
df
```

```python
df.fillna('', inplace=True)
df.drop(df.index[5:9], inplace=True)
df.drop(df.index[-3:], inplace=True)
df
```

```python
df['Petrol'] = df['Petrol'].str.replace(' ','')
df['Electric'] = df['Electric'].str.replace(' ','')
df['Diesel'] = df['Diesel'].str.replace(' ','')
df['Other(4)'] = df['Other(4)'].str.replace(' ','')
df['Sub-total'] = df['Sub-total'].str.replace(' ','')
df['Year'] = df['Year'].str.replace('/','')
df
```

```python
df.drop(['Petrol', 'Diesel', 'Other(4)'], axis=1, inplace=True)
df
```

```python
df["Date"] = df["Year"] + df["Month"]
df.drop(['Year', 'Month'], axis=1, inplace=True)
df = df.reindex(columns=['Date','Electric','Sub-total'])
df
```

```python
df.at[22,'Date']='Mar/21'
df.at[23,'Date']='Apr/21'
df.at[24,'Date']='May/21'
df.at[25,'Date']='Jun/21'
df.at[26,'Date']='Jul/21'
df.at[27,'Date']='Aug/21'
df.at[28,'Date']='Sep/21'
df.at[29,'Date']='Oct/21'
df.at[30,'Date']='Nov/21'
df.at[31,'Date']='Dec/21'
df.at[33,'Date']='Jan/22'
df.at[34,'Date']='Feb/22'
df.at[35,'Date']='Mar/22'
df.at[36,'Date']='Apr/22'
df.at[37,'Date']='May/22'
df.at[38,'Date']='Jun/22'
df.at[39,'Date']='Jul/22'
df.at[40,'Date']='Aug/22'
df.drop([32], axis=0, inplace=True)
df
```

```python
df[['Electric','Sub-total']] = df[['Electric','Sub-total']].astype(int)
df
```

```python
df['Sales share'] = (df['Electric'] / df['Sub-total']) *100
df
```

```python
df['Sales share (Cars using other fuels)'] = 100 - df['Sales share']
df
```

```python
df.drop(['Electric','Sub-total'], axis=1, inplace=True)
df
```

```python
plt.rc('figure', figsize=(50, 20))
ax = df.plot(x='Date', kind='bar', stacked=True, color=['#33CCCC', '#FFCC00'])
plt.xlabel('Date', fontsize=30)
plt.ylabel('Sales percentage', fontsize=30)
plt.title('Sales Percentage of Electric Cars and Gasoline Cars Over Years', fontsize=30)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.yaxis.set_minor_locator(plt.MultipleLocator(5))
```
